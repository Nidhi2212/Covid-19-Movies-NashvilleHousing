```python
# Importing libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = 12,8
```

```python
# Reading the data in the csv file

df = pd.read_csv(r'C:\Users\Nidhi\OneDrive\Desktop\Data Analytics\Project 1 - Guided (Alex)\movies.csv')
```

```python
#To display all the rows in the data

pd.set_option('display.max_rows', None)
```

```python
#Looking at the data, first 5 rows

df.head()
```

```python
# Data Cleaning

# To check if there is any missing value in the data

for col in df.columns:
    pct_missing_value = np.mean(df[col].isnull())
    print ('{} - {}%'.format(col, round(pct_missing_value*100,1)))
```

```python
#Datatypes of the columns

df.dtypes
```

```python
#Change data type of 2 columns

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')

df
```

```python
#Create a new column of Year from released column

New = df['released'].astype(str).str.split(",", n = 1, expand = True)

df['Released_Year'] = New[1].str[:5]

df
```

```python
#Sorting the data by gross column

df = df.sort_values(by=['gross'], inplace=False, ascending=False)
df
```

```python
# See unique values in company column

df['company'].drop_duplicates().sort_values(ascending = False)
```

```python
#To check the correlation between budget and gross by creating a scatter plot

plt.scatter(x = df['budget'], y = df['gross'])

plt.title('Budget vs Gross')

plt.xlabel('Gross_Earnings')

plt.ylabel('Film_Budget')

plt.show()
```

```python
#Creating line and scatter plot using seasborn and plotting gross and budget

sns.regplot(x = 'budget', y = 'gross', data = df, scatter_kws = {'color' : 'Red'}, line_kws = {'color' : 'blue'})
```

```python
# Look at the correlation between all numeric columns in the dataframe

#using Pearson method
print(df.corr(method='pearson'))

#using kendall method
print(df.corr(method='kendall'))

#using spearman method
print(df.corr(method='spearman'))
```

```python
correlation_matrix = df.corr()

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation b/w Numeric Data')

plt.xlabel('Movies Data')
plt.ylabel('Movies Data')

plt.show()
```

```python
#Convert the columns with object data type to numerical for including them into correlation matrix

df_categorical = df

for col_name in df_categorical.columns:
    if (df_categorical[col_name].dtype == 'object'):
        df_categorical[col_name] = df_categorical[col_name].astype('category')
        df_categorical[col_name] = df_categorical[col_name].cat.codes

df_categorical
```

```python
df
```

```python
#Creating a heatmap oout of correlation data of all the columns

correlation_matrix = df_categorical.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
#Looking at the Correlation unstacked for each of the columns

Correlation_Table = df_categorical.corr()

Correlation_Table_Unstacked = Correlation_Table.unstack()
Correlation_Table_Unstacked
```

```python
sorted_Correlation_Table_Unstacked = Correlation_Table_Unstacked.sort_values()
sorted_Correlation_Table_Unstacked
```

```python
High_Correlation = sorted_Correlation_Table_Unstacked[(sorted_Correlation_Table_Unstacked) > 0.5]
High_Correlation
```
